```python
compare_countries = final_data.groupby(["country", "region"]).agg(np.sum).reset_index()
```

```python
population = pd.read_excel("../data/socioeconomic_indicators_pop_growth.xlsx").loc[3:][["Citation:","Unnamed: 2"]]
gdp = pd.read_excel("../data/socioeconomic_indicators_GDP.xlsx").loc[3:][["Citation:","Unnamed: 7"]]

population = population.rename(index=str, columns={"Citation:": "country", "Unnamed: 2": "population"})
gdp = gdp.rename(index=str, columns={"Citation:": "country", "Unnamed: 7": "gdp"})
```

```python
del compare_countries['population']
```

```python
compare_countries = pd.merge(compare_countries, gdp, how='inner', on="country")
compare_countries = pd.merge(compare_countries, population, how='inner', on="country")
```

```python
compare_countries.head()
```

```python
compare_countries["gdp"] = np.log(pd.to_numeric(compare_countries["gdp"]))
compare_countries["population"] = np.log(pd.to_numeric(compare_countries["population"]))
```

```python
total = np.sum(compare_countries[['Global Covenant of Mayors',
       'EUCovenantofMayors2018', 'CarbonnApr2018', 'CDPCities2017',
       'Under2Coalition2016', 'WeAreStillIn', 'C40Cities2017', 'ClimateMayors',
       'CompactofStatesRegions2017', 'NCIReport',
       'CastanBroto_China_initiatives', 'IGDP2018', 'ClimateAlliance']], axis = 1)
```

```python
compare_countries["Initiatives Committed"] = np.log(total)
```

```python
compare_countries.head()
```

```python
np.unique(compare_countries.region)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np
data = []
regions = np.unique(compare_countries.region)
count = 0
for region in regions:
    df = compare_countries[compare_countries.region==region]

    x = df["gdp"]
    y = df["population"]
    z = df["Initiatives Committed"]
    trace1 = go.Scatter3d(name=region,
        x=x,
        y=y,
        z=z,
        mode='markers',
        marker=dict(
            size=12,
            line=dict(
                color='rgba(217, 217, 217, 0.14)',
                width=0.5
            ),
            opacity=0.8
        ),
        text=df["country"]
    )

    data += [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    ),
    scene = dict(
    xaxis = dict(
        title='GDP (Log)'),
    yaxis = dict(
        title='Population (Log)'),
    zaxis = dict(
        title='Initiatives Committed'),
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='simple-3d-scatter')
```
