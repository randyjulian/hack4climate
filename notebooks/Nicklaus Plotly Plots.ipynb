```python
import pandas as pd
import numpy as np
overshoot = pd.read_csv("../data/earthovershoot.csv")
```

```python
"2015-02-17"
dictionary = {"December":12,"November":11,"October":10, "September": 9, "August":8}
```

```python
list_of_dates = []
for raw_date in overshoot["Raw_Date"].tolist():
    year = raw_date[-4:]
    day = raw_date[-8:-6]
    month_num = -1
    if day[0] == " ":
        day = "0" + day[1:]
    for month in dictionary:
        if month in raw_date:
            month_num = dictionary[month]
    list_of_dates.append(year + "-" + str(month_num) + "-" + day)
```

```python
list_of_dates
```

```python
overshoot["Date"] = list_of_dates
```

```python
overshoot
```

```python
x = pandas.DatetimeIndex(start='2014-01-01', end='2014-01-06', freq='30T')
df = pandas.DataFrame(index=x, columns=['time since'])

basedate = pandas.Timestamp('2011-11-25')
df['time since'] = df.apply(lambda x: (x.name.to_datetime() - basedate).days, axis=1)
```

```python
overshoot_days = []
for date in list_of_dates:
    day = pd.Timestamp(date) - pd.Timestamp(date[:4]+"-01-01")
    overshoot_days.append(day.days)
```

```python
overshoot["Overshoot Days"] = overshoot_days
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

data = [go.Scatter(
          x=overshoot.Date,
          y=overshoot['Overshoot Days'])]

py.iplot(data, filename = "Earth Overshoot Days")
```

```python
compare_countries = pd.read_csv("final_data.csv")
compare_countries.head()
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np

data = []
regions = np.unique(compare_countries.region)
count = 0
for region in regions:
    df = compare_countries[compare_countries.region==region]

    x = df["gdp"]
    y = df["population"]
    z = np.log(df["Initiatives Committed"])
    trace1 = go.Scatter3d(name=region,
        x=x,
        y=y,
        z=z,
        mode='markers',
        marker=dict(
            size=12,
            line=dict(
                color='rgba(217, 217, 217, 0.14)',
                width=0.5
            ),
            opacity=0.8
        ),
        text=df["country"]
    )

    data += [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    ),
    scene = dict(
    xaxis = dict(
        title='GDP (Log)'),
    yaxis = dict(
        title='Population (Log)'),
    zaxis = dict(
        title='Initiatives Committed (Log)'),
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='simple-3d-scatter')
```
